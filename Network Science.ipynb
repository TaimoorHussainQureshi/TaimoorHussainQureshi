```python
# f) Removing edges based on link betweenness
for _ in range(5):
    edges_sorted_by_betweenness = sorted(link_betweenness.items(), key=lambda x: x[1])
    half_edges = edges_sorted_by_betweenness[:len(edges_sorted_by_betweenness)//2]
    
    G.remove_edges_from([edge for edge, _ in half_edges])
    link_betweenness = nx.edge_betweenness_centrality(G)
    connected_components = nx.number_connected_components(G)
    print("Number of connected components:", connected_components)
```

```python
# g) Repeating with decreasing link betweenness
for _ in range(5):
    edges_sorted_by_betweenness = sorted(link_betweenness.items(), key=lambda x: x[1], reverse=True)
    half_edges = edges_sorted_by_betweenness[:len(edges_sorted_by_betweenness)//2]
    
    G.remove_edges_from([edge for edge, _ in half_edges])
    link_betweenness = nx.edge_betweenness_centrality(G)
    connected_components = nx.number_connected_components(G)
    print("Number of connected components:", connected_components)
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# a) Analyzing Random Networks

# Calculate ensemble average degree ⟨𝑘⟩
N = 10001
p = 0.01
avg_degree = N * p

# Approximate probability for degree range
prob_range = (avg_degree, 2 * avg_degree)

# Generate a G(N, p) network
G_random = nx.erdos_renyi_graph(N, p)

# Calculate average degree 𝑘 for G_random
k_values = [deg for _, deg in G_random.degree()]
k_avg = np.mean(k_values)

# Calculate probability for degree range for G_random
prob_k_range = sum(1 for k in k_values if prob_range[0] < k < prob_range[1]) / N

avg_degree
prob_k_range
prob_range
```

```python
# b) Watts-Strogatz Networks

pd_values = np.arange(0, 1.01, 0.01)
avg_distances = []
avg_clustering_coeffs = []

for pd in pd_values:
    # Generate 5 Watts-Strogatz networks
    networks = [nx.watts_strogatz_graph(500, 6, pd) for _ in range(5)]

    # Calculate average distance 𝑑 and average clustering coefficient 𝐶
    distances = [nx.average_shortest_path_length(net) for net in networks]
    clustering_coeffs = [nx.average_clustering(net) for net in networks]

    avg_distances.append(np.mean(distances))
    avg_clustering_coeffs.append(np.mean(clustering_coeffs))

# Plot average distance 𝑑 and average clustering coefficient 𝐶
plt.figure(figsize=(10, 6))
plt.plot(pd_values, avg_distances, label='Average Distance')
plt.plot(pd_values, avg_clustering_coeffs, label='Average Clustering Coefficient')
plt.xlabel("Rewiring Probability (p𝑑)")
plt.ylabel("Value")
plt.title("Watts-Strogatz Networks Analysis")
plt.legend()
plt.show()
```

```python
# c) Windmill Graph and Barabasi-Albert Network

# Construct the windmill graph Wd(k, n) and display the graph
def construct_windmill_graph(k, n):
    G_windmill = nx.Graph()
    for i in range(n):
        for j in range(i + 1, n):
            for l in range(k):
                G_windmill.add_edge((i, l), (j, l))
    return G_windmill

# Display the windmill graph
G_windmill = construct_windmill_graph(k=5, n=3)
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_windmill)
nx.draw(G_windmill, pos, with_labels=False, node_size=100)
plt.title("Windmill Graph Wd(5, 3)")
plt.show()
```

```python
# Generate a Barabasi-Albert network with initial configuration Wd(5, 3)
G_BA = nx.barabasi_albert_graph(8000, 5)

# Compute degree distribution for the original Barabasi-Albert network
degree_dist_BA = np.array([deg for _, deg in G_BA.degree()])
unique_degrees, degree_counts = np.unique(degree_dist_BA, return_counts=True)
degree_probs_BA = degree_counts / np.sum(degree_counts)
G_BA
```

```python
# Plot the degree distribution for the original Barabasi-Albert network
plt.figure(figsize=(10, 6))
plt.scatter(unique_degrees, degree_probs_BA, s=50, label='Barabasi-Albert (Original)')
plt.yscale('log')
plt.xscale('log')
plt.xlabel("Degree (log scale)")
plt.ylabel("Probability (log scale)")
plt.title("Degree Distribution of Barabasi-Albert Network (Original)")
plt.legend()
plt.show()
```
